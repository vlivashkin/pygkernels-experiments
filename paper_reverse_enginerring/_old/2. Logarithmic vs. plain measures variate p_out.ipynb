## 2. Logarithmic vs. plain measures
Let $G(N,(m)p_{in}, p_{out})$ be the model of generating random graphs on $N$ nodes divided into $m$ classes of the same size, with $p_{in}$ and pout being the probability of $(i, j) \in E(G)$ for $i$ and $j$ belonging to the same class and different classes, respectively, where $E(G)$ is the edge set of $G$.

The curves in Figures 1–3 present the adjusted Rand index (averaged over 200 random graphs) for clustering with Ward’s
method.

```python
def plot_log_results(results):
    fig, ax = plt.subplots(1, 4, figsize=(15, 4))
    for idx_i, (name1, name2, name3, name4, xlim, ylim) in enumerate([
        [(f'KernelKMeansSklearn_pWalk H', 'kMeans pWalk'),
         (f'KernelKMeansSklearn_Walk H', 'kMeans Walk'),
         (f'KernelWardSklearn_pWalk H', 'Ward pWalk'),
         (f'KernelWardSklearn_Walk H', 'Ward Walk'),
         (0, 1), (0.0, 1)],
        [(f'KernelKMeansSklearn_For H', 'kMeans For'),
         (f'KernelKMeansSklearn_logFor H', 'kMeans logFor'),
         (f'KernelWardSklearn_For H', 'Ward For'),
         (f'KernelWardSklearn_logFor H', 'Ward logFor'),
         (0, 1), (0.0, 1)],
        [(f'KernelKMeansSklearn_Comm H', 'kMeans Comm'),
         (f'KernelKMeansSklearn_logComm H', 'kMeans logComm'),
         (f'KernelWardSklearn_Comm H', 'Ward Comm'),
         (f'KernelWardSklearn_logComm H', 'Ward logComm'),
         (0, 0.6), (0.0, 1)],
        [(f'KernelKMeansSklearn_Heat H', 'kMeans Heat'),
         (f'KernelKMeansSklearn_logHeat H', 'kMeans logHeat'),
         (f'KernelWardSklearn_Heat H', 'Ward Heat'),
         (f'KernelWardSklearn_logHeat H', 'Ward logHeat'),
         (0, 0.9), (0.0, 1)]
    ]):
        toplot = [
            (name1[1], *results[name1[0]]),
            (name2[1], *results[name2[0]]),
            (name3[1], *results[name3[0]]),
            (name4[1], *results[name4[0]]),
        ]
        plot_results(ax[idx_i], toplot, xlim, ylim)
```

### **Fig. 1** Logarithmic vs. plain measures for G(100,(2)0.2,0.05)

```python
@load_or_calc_and_save('results/2_11_ari_short.pkl')
def calc():
    results = defaultdict(list)
    graphs, info = StochasticBlockModel(100, 2, p_in=0.2, p_out=0.05).generate_graphs(200)
    classic_plot = ParallelByGraphs(adjusted_rand_score, np.linspace(0, 1, 101), progressbar=True)
    for clustring_method in tqdm([KernelKMeansSklearn, KernelWardSklearn]):
        for measure_class in tqdm([pWalk_H, Walk_H, For_H, logFor_H, Comm_H, logComm_H, Heat_H, logHeat_H]):
            results[f"{clustring_method.name}_{measure_class.name}"] = \
                classic_plot.perform(clustring_method, measure_class, graphs, 2, n_jobs=12)
    return results

results = calc()
```

```python
plot_log_results(results)
# gcf().suptitle("Fig. 1 Logarithmic vs. plain measures for G(100,(2)0.2,0.05)");
```
