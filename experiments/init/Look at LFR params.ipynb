```python
import sys
import pandas as pd

sys.path.append("../../pygkernels")
from pygkernels.data import Datasets, LFRGenerator
```

```python
datasets_names = [
    'dolphins',
    'football',
    'karate',
    'polbooks',
    'sp_school_day_1', 'sp_school_day_2',
    'news_2cl1_0.1', 'news_2cl2_0.1', 'news_2cl3_0.1',
    'news_3cl1_0.1', 'news_3cl2_0.1', 'news_3cl3_0.1',
    'news_5cl1_0.1', 'news_5cl2_0.1', 'news_5cl3_0.1',
    'polblogs',
    'cora_DB', 'cora_EC', 'cora_HA',
    'cora_HCI', 'cora_IR', 'cora_Net',
    'eu-core',
    'eurosis'
]
```

```python
results = []
datasets = Datasets()
for dataset_name in datasets_names:
    (A, partition), info = datasets[dataset_name]
    gen = LFRGenerator.params_from_adj_matrix(A, partition, info['k'], name=dataset_name)
    results.append(gen.generate_info())
```

```python
df = pd.DataFrame(results, columns=['name', 'n', 'tau1', 'tau2', 'mu', 'average_degree', 'max_degree', 'min_community', 'max_community'])
df.to_excel('lfr_params.xlsx')
df
```
